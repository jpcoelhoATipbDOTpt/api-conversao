#**Configurações no Eclipse**

##**Iniciando um projeto no Eclipse**



*   Abrir o Eclipse;
*   File>New>Java Project;
*   Insira o nome do projeto (de preferência em "CamelCase");
*   No campo "JRE";
*   Em "Use an execution environment JRE:" selecione "JavaSE-1.8";
*   Clicar em "Next" depois "Finish" mantendo as configurações padrões;
*   Clicar com o botão direito no projeto;
*   Properties>"Java Build Path">"Add External JARs...";
*   Selecionar o arquivo "jade.jar" que foi extraído do ficheiro inicialmente;
*   "Apply and Close";





##**Iniciando uma classe no Eclipse**

*   No "Package Explorer" (aba lateral do Eclipse), abra o projeto e clique com o botão direito na pasta "src";
*   "New">"Class";
*   Manter o package como default;
*   Escrever o nome da classe;
*   "Finish".


##**Rodar um código JADE no Eclipse**

*   Iniciar o JADE;
*   java jade.Boot -gui;
*   Caso não funcione, utilize o comando java -classpath "caminho do arquivo jade.jar" jade.Boot -gui;
*   Clicar na seta ao lado do "Run" (botão play verde);
*   "Run Configurations";
*   Colocar um nome para a configuração e o nome do projeto a ser inicializado;
*   Na aba "Arguments">"Program arguments";
*   Colocar o comando: -container NomeDoAgente:NomeDaClasse(argumentos);
*   O nome da classe deve ser exatamente igual ao nome da classe com o código;
*   "Apply">"Run".
*   Se ao rodar não aparecer a informação desejada, verifique na tela do JADE se foi encerrado o último agente.

#**HelloWorld**

Criar um agente JADE que imprime na tela a frase Hello World. Agentes JADE são definidos como subclasses da classe Agent e seu código inicial (o que o agente fará ao ser executado) deve ser colocado dentro do método setup().

```python
import jade.core.Agent;

public class HelloWorld extends Agent {
	
	protected void setup() {
		// Executa as tarefas de inicialização
		System.out.println("Hello World!");
		System.out.println("Meu nome é " + getLocalName());
	}
}
```

Execução:

-container a1:HelloWorld

###**HelloWorldInfo**

De acordo com o padrão FIPA cada instância de agente é identificada por um agent identifier.

```python
import java.util.Iterator;

import jade.core.Agent;

public class HelloWorldInfo extends Agent {
	
	protected void setup() {
		System.out.println("Hello World! Eu sou um agente!");
		System.out.println("Todas as minhas informações: \n" + getAID()); // O método getAID() permite obter a identificação completa do agente
		System.out.println("Meu nome local é " + getAID().getLocalName()); // Obtém o nome do agente na plataforma
		System.out.println("Meu nome global (GUID) é " + getAID().getName()); // Obtém o nome global
		System.out.println("Meus endereços são: ");
		Iterator it = getAID().getAllAddresses();
		while(it.hasNext()) { // Percorre a lista de endereços
			System.out.println("- " + it.next());
		}
	}
}
```
